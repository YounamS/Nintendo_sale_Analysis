# importing all the required library and data file

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\C0749068\Desktop\dataanylst\vgsales.csv')
df
```

# Droping all the null values

```python
df = df.dropna()
df
```

# sorting all values by year 

```python
df = df.sort_values('Year')
df
```

# filtering all value to games which are published by nintendo 

```python
df = df.query('Publisher == "Nintendo"')
df
```

# sales through out the year 1983 to 2016

```python
sales = df.groupby('Year',as_index=False).sum()
sales = sales.drop(columns=['Name','Rank','Platform','Publisher','Genre'])
sales
```

```python

y1 = np.array(sales["Global_Sales"])
y2 = np.array(sales["JP_Sales"])
y3 = np.array(sales["EU_Sales"])
y4 = np.array(sales["NA_Sales"])
y5 = np.array(sales["Other_Sales"])
x1 = np.array(sales["Year"])

plt.plot(x1,y1,x1,y2,x1,y3,x1,y4,x1,y5)

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title("Sales by year")
plt.legend()
```

# Number of games released each year 

```python
#game relased by year
g_relased =df.groupby(['Year'],as_index=False).count()
g_relased =g_relased.drop(columns=['Name','Platform','Publisher','Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales'])
g_relased =g_relased.rename(columns={'Rank':'Game_Released'})
g_relased
```

```python
g_relased.plot(x="Year",y="Game_Released",kind='bar')
```

# game released of each genre between year 1983 to 2016

```python
# genre count by year
g_count=df.groupby(['Year','Genre'],as_index=False).count()
g_count =g_count.drop(columns=['Name','Platform','Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales'])
g_count=g_count.rename(columns={'Rank':'Genre_count'})
g_count
```

```python
pie = g_count.pivot_table(index='Genre',aggfunc='size')
pie
```

```python
y =np.array(pie)
y1 = np.array(pie)

plt.pie(y1,labels=y)
plt.show()
```
